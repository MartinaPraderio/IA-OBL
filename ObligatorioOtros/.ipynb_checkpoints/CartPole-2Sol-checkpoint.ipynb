```python
import gym
import numpy as np
```

```python
env = gym.make('CartPole-v1')
```

```python
def epsilon_greedy_policy(state, Q, epsilon=0.1):
    explore = np.random.binomial(1, epsilon)
    if explore:
        action = env.action_space.sample()
        #print('explore')
    else:
        action = np.argmax(Q[state])
        #print('exploit')
    return action
```

```python
def optimal_policy(state, Q):
    action = np.argmax(Q[state])
    return action
```

```python
bins_position = np.linspace(-5, 5, 10)
bins_angle = np.linspace(-0.418 , 0.418, 10)
bins_inf = np.linspace(-100 , 100, 10)
```

```python
def get_state(obs):
    d_position = int(np.digitize(obs[0], bins_position))
    d_angle = int(np.digitize(obs[2], bins_angle))
    d_angularVelocity = int(np.digitize(obs[3], bins_inf))
    d_cartVelocity = int(np.digitize(obs[1], bins_inf))
    state = (d_position, d_cartVelocity, d_angle, d_angularVelocity)
    return state
```

```python
state = get_state([-1.4, -2., -0.1, 1.2])
state
```

```python
Q = np.random.random((10,10,10,10,2))
print(Q)
```

```python
gamma = 0.5
alpha = 0.5
for i in range (100000):
    obs = env.reset()
    #print(obs)
    done = False
    while not done:
        state = get_state(obs)
        action = epsilon_greedy_policy(state, Q, 0.5)
        obs, reward, done, info = env.step(action)
        update_Q(Q, state, action, reward, get_state(obs), alpha, gamma)
        #print('->', state, action, reward, obs, done, info)
# env.close()
```

```python
def update_Q(Q , state, action, reward, next_state, alpha, gamma):
    Q[state][action] = Q[state][action] + alpha*(reward + gamma*(optimal_policy(next_state, Q)) - Q[state][action])
```
